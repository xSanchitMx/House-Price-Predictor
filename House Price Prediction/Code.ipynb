```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("California Housing Prices/housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis=1)
y = data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)   
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data_corr = train_data.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data_corr, annot=True,cmap="Reds")
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
dummy = pd.get_dummies(train_data.ocean_proximity)
train_data = train_data.join(dummy)
train_data = train_data.drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
train_data_corr = train_data.corr()
plt.figure(figsize=(15,8))
sns.heatmap(train_data_corr, annot=True,cmap="Purples")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
train_data_corr = train_data.corr()
plt.figure(figsize=(15,8))
sns.heatmap(train_data_corr, annot=True,cmap="Reds")
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train=train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg=LinearRegression()

reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

dummy = pd.get_dummies(test_data.ocean_proximity)
test_data = test_data.join(dummy)
test_data = test_data.drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']


```

```python
x_test, y_test=test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest=RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test,y_test)
```

```python
y_pred=forest.predict(x_test)
performance = pd.DataFrame({'PREDICTIONS': y_pred, 'ACTUAL VALUES':y_test})
performance['error']=performance['ACTUAL VALUES']-performance['PREDICTIONS']
```

```python
performance
```

```python
performance.reset_index(drop=True, inplace=True) #inplace turns into a column
performance.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar('index', 'error', data=performance[:50], color='black', width=0.3)
plt.ylabel('residuals')
plt.xlabel('observations')
plt.show()
```
