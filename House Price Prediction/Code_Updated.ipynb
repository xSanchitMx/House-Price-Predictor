```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("California Housing Prices/housing.csv")
```

```python
data
```

```python
data.info()
```

```python
#There are some null datas in total_bedrooms
data=data.dropna()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.hist(bins=50,figsize=(20,15))
```

```python
dummy = pd.get_dummies(data.ocean_proximity)
data = data.join(dummy)
data = data.drop(['ocean_proximity'], axis=1)
```

```python
data
```

```python
data_corr=data.corr()
```

```python
data_corr
```

```python
data_corr["median_house_value"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data_corr, annot=True,cmap="Blues")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=data, hue="median_house_value", palette="coolwarm")
```

```python
x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, 
                                                    test_size = 0.2, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
LR=LinearRegression()
LR.fit(x_train, y_train)
```

```python
LR.score(x_train, y_train)
```

```python
y_pred_LR=LR.predict(x_test)
performance_LR=pd.DataFrame({'PREDICTIONS': y_pred_LR, 'ACTUAL VALUES':y_test, 'ERROR': y_pred_LR-y_test})
```

```python
performance_LR
```

```python
performance_LR.reset_index(drop=True, inplace=True)
performance_LR.reset_index(inplace=True)
```

```python
performance_LR
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar('index', 'ERROR', data=performance_LR[:100], color='black', width=0.3)
plt.ylabel('residuals')
plt.xlabel('observations')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(x_train, y_train)
```

```python
forest.score(x_train, y_train)
```

```python
y_pred_forest=forest.predict(x_test)
performance_forest=pd.DataFrame({'PREDICTIONS': y_pred_forest, 'ACTUAL VALUES': y_test, 'ERROR': y_pred_forest-y_test})
```

```python
performance_forest
```

```python
performance_forest.reset_index(drop=True, inplace=True)
performance_forest.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar('index', 'ERROR', data=performance_forest[:100], color='black', width=0.3)
plt.ylabel('residuals')
plt.xlabel('observations')
plt.show()
```

```python
performance_forest.drop(['index'],axis=1,inplace=True)
performance_forest.to_excel('predictions.xlsx')
```
